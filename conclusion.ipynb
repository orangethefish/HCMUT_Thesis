```python
from matplotlib import pyplot as plt
```

```python
model_names = ['old_model', 'new_model', 'data_augmentation_model', 'transfer_learning_model', 'synthetic_model']
model_accuracies = [0.61, 0.82, 0.92, 0.84, 0.80]

plt.bar(model_names, model_accuracies)
plt.xlabel('Model Name')
for i in range(len(model_accuracies)):
    plt.text(i, model_accuracies[i], model_accuracies[i], ha = 'center')
plt.ylabel('Model Accuracy')
plt.title('Model Accuracies')
plt.xticks(rotation=45) 
plt.show()
```
